```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data
```

```python
data_1 = data.copy()
data_1.loc[data_1['whoAmI'] == 'human', 'human'] = 1
data_1.loc[data_1['whoAmI'] == 'robot', 'robot'] = 1
data_1.pop('whoAmI')
data_1.fillna(value = 0, inplace = True)
data_1 = data_1.astype(int)
data_1
```
